```python
import datetime
import pandas as pd
```

```python
df = pd.read_csv('crypto/newer_old_cryptodata.csv', usecols=['timestamp', 'last'])
df
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'],unit='s')
```

```python
df = df.set_index('timestamp')
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', 3):
    print(df)
```

```python
df = df.resample('1H').first()
df
```

```python
df.drop(df.index[[0,95]], inplace=True)
df
```

```python
df.to_csv('crypto/processed_may5may9.csv')
```
